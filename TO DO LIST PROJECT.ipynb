```python
import tkinter as tk
from tkinter import messagebox

# Create the main application window
root = tk.Tk()
root.title("To-Do List Application")
root.geometry("400x400")  # Set the window size

# Create a list to store tasks
to_do_list = []

# Function to add a task
def add_task():
    task = entry.get()
    if task:
        to_do_list.append(task)
        entry.delete(0, tk.END)  # Clear the entry field
        update_list()
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

# Function to remove a task
def remove_task():
    selected_task = listbox.get(tk.ACTIVE)
    if selected_task in to_do_list:
        to_do_list.remove(selected_task)
        update_list()

# Function to edit a task
def edit_task():
    selected_task = listbox.get(tk.ACTIVE)
    if selected_task in to_do_list:
        new_task = entry.get()
        if new_task:
            index = to_do_list.index(selected_task)
            to_do_list[index] = new_task
            entry.delete(0, tk.END)  # Clear the entry field
            update_list()
        else:
            messagebox.showwarning("Warning", "Please enter a new task.")

# Function to display the to-do list
def update_list():
    listbox.delete(0, tk.END)  # Clear the listbox
    for task in to_do_list:
        listbox.insert(tk.END, task)

# Create labels and entry field
label = tk.Label(root, text="Enter a Task:", font=("Arial", 14))
label.pack(pady=10)

entry = tk.Entry(root, font=("Arial", 14))
entry.pack()

# Create buttons with different colors and larger size
add_button = tk.Button(root, text="Add Task", command=add_task, bg="green", font=("Arial", 14))
add_button.pack(pady=10)

remove_button = tk.Button(root, text="Remove Task", command=remove_task, bg="red", font=("Arial", 14))
remove_button.pack()

edit_button = tk.Button(root, text="Edit Task", command=edit_task, bg="blue", font=("Arial", 14))
edit_button.pack()

# Create a listbox to display tasks
listbox = tk.Listbox(root, selectmode=tk.SINGLE, font=("Arial", 14))
listbox.pack(pady=10)

# Create an initial display of the to-do list
update_list()

root.mainloop()
```

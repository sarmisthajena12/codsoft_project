```python
from tkinter import *
import math

# Global variable for input
i = 0

# Create the main application window
root = Tk()
root.title("Simple Calculator")
root.geometry("700x800")

# Create an input field (Entry widget)
display = Entry(root)
display.grid(row=0, column=0, columnspan=6, padx=10, pady=10, ipadx=10, ipady=10)

# Function to update the display
def update_display(value):
    current_text = display.get()
    display.delete(0, END)
    display.insert(0, current_text + str(value))

# Function to clear the display
def clear_display():
    display.delete(0, END)

# Function to perform the calculation
def calculate():
    try:
        expression = display.get()
        result = str(eval(expression))
        clear_display()
        display.insert(0, result)
    except Exception as e:
        clear_display()
        display.insert(0, "Error")

# Function to calculate factorial
def calculate_factorial():
    try:
        number = int(display.get())
        result = math.factorial(number)
        clear_display()
        display.insert(0, result)
    except Exception as e:
        clear_display()
        display.insert(0, "Error")

# Create buttons with different colors and larger size
buttons = [
    "7", "8", "9", "+",
    "4", "5", "6", "-",
    "1", "2", "3", "*",
    "C", "0", "=", "/",
    "pi", "%", "(", ")",
    "exp", "^2", "<-", "x!"
]

row_num, col_num = 1, 0

for button in buttons:
    if button == "=":
        Button(root, text=button, command=calculate, bg="green", font=("Arial", 20)).grid(row=row_num, column=col_num, sticky=N + S + E + W)
    elif button == "C":
        Button(root, text=button, command=clear_display, bg="red", font=("Arial", 20)).grid(row=row_num, column=col_num, sticky=N + S + E + W)
    elif button == "x!":
        Button(root, text=button, command=calculate_factorial, bg="lightblue", font=("Arial", 20)).grid(row=row_num, column=col_num, sticky=N + S + E + W)
    else:
        Button(root, text=button, command=lambda b=button: update_display(b), bg="lightblue", font=("Arial", 20)).grid(row=row_num, column=col_num, sticky=N + S + E + W)

    col_num += 1
    if col_num > 3:
        col_num = 0
        row_num += 1

# Start the main event loop
root.mainloop()
```
